````python
from tqdm import tqdm
````

````python
import sys
import yaml
from pathlib import Path
from easydict import EasyDict
import os

from pcdet.datasets import KittiDataset

import pandas_profiling
import pandas as pd
import numpy as np
````

````python
# For Carla-Kitti
# DATASET_NAME = 'carla-kitti'
# DATASET_VERSION = '0.0'
# ds_name_ver = DATASET_NAME + '_v' + DATASET_VERSION

# For Kitti
ds_name_ver = 'kitti'
DATASET_NAME = 'kitti'

os.mkdir(ds_name_ver)
````

````python
dataset_cfg_dir = '/home/farzad/PycharmProjects/OpenPCDet/tools/cfgs/dataset_configs/kitti_dataset.yaml'
pcdet_root_dir = Path('/home/farzad/PycharmProjects/OpenPCDet/')
class_names = ['Pedestrian', 'Car']
````

````python
dataset_cfg = EasyDict(yaml.load(open(dataset_cfg_dir)))
data_path = pcdet_root_dir / 'data' / DATASET_NAME
save_path = pcdet_root_dir / 'data' / DATASET_NAME
````

````python
kitti = KittiDataset(dataset_cfg=dataset_cfg, class_names=class_names, root_path=data_path)
````

````python
infos = kitti.get_infos();
````

````python
per_img_infos = []
per_obj_infos = []

for info in tqdm(infos):
    
    annos = info['annos']
    num_gts = annos['gt_boxes_lidar'].shape[0]
    
    lidar_idx = info['point_cloud']['lidar_idx']
    
    img_info = {}
    img_info['lidar_idx'] = lidar_idx
    img_info['num_gts'] = num_gts
    img_info['img_height'] = info['image']['image_shape'][0]
    img_info['img_width'] = info['image']['image_shape'][1]
        
    for i in range(num_gts):
        obj_info = {}
        obj_info['lidar_idx'] = lidar_idx
        
        
        for key, val in annos.items():
            if val.ndim == 2:  # split columns with 2D elements
                for j in range(val.shape[-1]):
                    new_key = key + f'_{j}'
                    obj_info[new_key] = val[i, j]
            else:
                obj_info[key] = val[i]
            
        per_obj_infos.append(obj_info)
    
    per_img_infos.append(img_info)
````

````python
per_img_infos_pd = pd.DataFrame.from_dict(per_img_infos)
per_obj_infos_pd = pd.DataFrame.from_dict(per_obj_infos)
new_colnames = ['lidar_idx', 'name', 'truncated', 'occluded', 'alpha', 'bbox_left',
                'bbox_top', 'bbox_right', 'bbox_bottom', 'dimensions_length',
                'dimensions_width', 'dimensions_height', 'location_x',
                'rotation_y', 'rotation_z', 'rotation_y', 'score', 'difficulty', 'index',
                'loc_x', 'loc_y', 'loc_z', 'dim_x', 'dim_y',
                'dim_z', 'rotation_y_lidar', 'num_points_in_gt']
per_obj_infos_pd.columns = new_colnames
per_obj_infos_pd = per_obj_infos_pd.drop(columns=['lidar_idx', 'index'])
height = per_obj_infos_pd['bbox_bottom'] - per_obj_infos_pd['bbox_top']
per_obj_infos_pd['pixel_height'] = height


report_img = pandas_profiling.ProfileReport(per_img_infos_pd, minimal=True)
report_obj = pandas_profiling.ProfileReport(per_obj_infos_pd, minimal=True)
````

````python
report_img.to_file(os.path.join(ds_name_ver, DATASET_NAME + '_per_img.html'))
report_obj.to_file(os.path.join(ds_name_ver, DATASET_NAME + '_per_obj.html'))
````
